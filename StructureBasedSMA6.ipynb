```python
import networkx as nx
import matplotlib.pyplot as plt

# Create graph
G = nx.Graph()
G.graph['name'] = "Mini Social Network"

# Indian user names (6 nodes)
users = ["Amit", "Priya", "Raj", "Sneha", "Anil", "Deepa"]
G.add_nodes_from(users)

# Add connections
connections = [
    ("Amit", "Priya"),
    ("Amit", "Raj"),
    ("Raj", "Sneha"),
    ("Sneha", "Anil"),
    ("Anil", "Deepa"),
    ("Deepa", "Amit"),
    ("Priya", "Deepa")  # Slightly more connected
]
G.add_edges_from(connections)

# Simple group coloring
group_1 = {"Amit", "Priya", "Deepa"}
group_2 = {"Raj", "Sneha", "Anil"}

color_map = []
for node in G.nodes:
    if node in group_1:
        color_map.append('lightblue')
    else:
        color_map.append('lightgreen')

# Centrality-based node sizing
degree_centrality = nx.degree_centrality(G)
node_size = [1000 + 3000 * degree_centrality[node] for node in G.nodes]

# Draw the graph
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G, seed=42)

nx.draw_networkx_nodes(G, pos, node_color=color_map, node_size=node_size)
nx.draw_networkx_edges(G, pos, width=2, alpha=0.5)
nx.draw_networkx_labels(G, pos, font_size=12, font_weight='bold')

plt.title("Mini Social Media Network (Manual Groups)", fontsize=14)
plt.axis('off')
plt.show()
```
